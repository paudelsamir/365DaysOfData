```python
# list of redundant features to drop
drop_columns = [
    "Total attacking", "Total skill", "Total defending", "Total power", "Total stats", 
    "Total mentality", "Total goalkeeping", 
    "Best position_CAM", "Best position_CB", "Best position_CDM", "Best position_CF", 
    "Best position_CM", "Best position_GK", "Best position_LB", "Best position_LM", 
    "Best position_LW", "Best position_LWB", "Best position_RB", "Best position_RM", 
    "Best position_RW", "Best position_RWB", "Best position_ST", 
]

df = df.drop(columns=drop_columns)
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Drop features
X_reduced = df.drop(columns=to_drop)

# Recalculate VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X_reduced.columns
vif_data["VIF"] = [variance_inflation_factor(X_reduced.values, i) for i in range(X_reduced.shape[1])]

print(vif_data)
```

```python
# Train a RandomForestRegressor to calculate feature importance
model_rf = RandomForestRegressor(random_state=42)
model_rf.fit(X_train, y_train)

# Get feature importances
importances = model_rf.feature_importances_

# Create a DataFrame for better visualization
feature_importances = pd.DataFrame({'Feature': X_train.columns, 'Importance': importances})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

print(feature_importances)
```

```python
# Plot the top 10 features
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importances.head(10))
plt.title('Top 10 Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
# Define the features and target variable
X = df.drop(columns=['Value'])
y = df['Value']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Train the Linear Regression model
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

# Predict on the test set
y_pred = model_lr.predict(X_test)

# Evaluate the model
print('R2 Score:', r2_score(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
```
