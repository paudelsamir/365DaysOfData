```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
X, y = make_moons(n_samples=500, noise=0.25, random_state=42)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
model1 = Sequential()
model1.add(Dense(64, input_dim=2, activation='relu'))
model1.add(Dense(64, activation='relu'))
model1.add(Dense(1, activation='sigmoid'))
```

```python
model1.summary()
```

```python
adam = Adam(learning_rate=0.01)
model1.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'] )
```

```python
history1 = model1.fit(X, y, validation_split= 0.2, epochs=100, verbose=1)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X, y.astype('int'), clf=model1, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

```python
model2 = Sequential()
model2.add(Dense(64, input_dim=2, activation='relu', kernel_regularizer= tensorflow.keras.regularizers.l2(0.01)))
model2.add(Dense(64, activation='relu', kernel_regularizer= tensorflow.keras.regularizers.l2(0.01)))
model2.add(Dense(1, activation='sigmoid'))
```

```python
model2.summary()
```

```python
adam = Adam(learning_rate=0.01)
model2.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'] )
```

```python
history2 = model2.fit(X, y, validation_split= 0.2, epochs=100, verbose=1)
```

```python
plot_decision_regions(X, y.astype('int'), clf=model2, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
```
