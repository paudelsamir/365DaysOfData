```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
```

# New Section

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
# Load an example image
(imgs, labels), _ = tf.keras.datasets.cifar10.load_data()
img = imgs[0]

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(img.astype('uint8'))
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Reshape the image to (1, height, width, channels)
img = img.reshape((1,) + img.shape)
```

```python
# Generate augmented images and display them
i = 0
augmented_images = []
for batch in datagen.flow(img, batch_size=1):
    augmented_images.append(batch[0].astype('uint8'))
    i += 1
    if i % 4 == 0:  # Generate 4 images
        break
```

```python
# Display augmented images in matrix format
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes = axes.flatten()
for img, ax in zip(augmented_images, axes):
    ax.imshow(img)
    ax.axis('off')
plt.suptitle("Augmented Images")
plt.show()
```

```python
# Print the matrix format of the augmented images
print("Original Image Matrix:")
print(imgs[0])

for idx, aug_img in enumerate(augmented_images):
    print(f"\nAugmented Image {idx + 1} Matrix:")
    print(aug_img)
```
