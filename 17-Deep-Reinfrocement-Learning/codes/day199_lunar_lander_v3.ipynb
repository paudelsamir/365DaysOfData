```python
import gymnasium

from huggingface_sb3 import load_from_hub, package_to_hub
from huggingface_hub import (
    notebook_login,
) 

from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.monitor import Monitor
```

```python
import gymnasium as gym
```

```python
env = gym.make("LunarLander-v3", render_mode="human")


# now we use the PPO model, combination of value based rl and policy based rl
model = PPO('MlpPolicy', env, verbose=1)


model.learn(total_timesteps=10000)
```

```python
# We create our environment with gym.make("<name_of_the_environment>")
env = gym.make("LunarLander-v3", render_mode="human")
env.reset()
print("_____OBSERVATION SPACE_____ \n")
print("Observation Space Shape", env.observation_space.shape)
print("Sample observation", env.observation_space.sample())  # Get a random observation

print("\n _____ACTION SPACE_____ \n")
print("Action Space Shape", env.action_space.n)
print("Action Space Sample", env.action_space.sample())  # Take a random action

env.close()
```

```python
env = gym.make("LunarLander-v3" ,render_mode = "rgb_array")
model = PPO(
    policy="MlpPolicy",
    env=env,
    n_steps=1024,
    batch_size=64,
    n_epochs=4,
    gamma=0.999,
    gae_lambda=0.98,
    ent_coef=0.01,
    verbose=1,
)
```

```python
model.learn(total_timesteps=1000000)
model_name = "ppo-LunarLander-sam"
model.save(model_name)
```

```python
eval_env = Monitor(gym.make("LunarLander-v3"))
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)
print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```

In my case, I got a mean_reward=226.56 +/- 78.56714748934611 after training for 1 million steps for 30 mins, which means that our lunar lander agent is ready to land on the moon
